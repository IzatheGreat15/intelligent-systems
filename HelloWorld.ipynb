<h5>Practical Example</h5>
<p>Let's try making a graph to represent the sales trajectory of a company within 4 quarters, given that they have two types of customers(male and female).</p>

<p>This time, let us use <i>numpy</i> to generate the data</p>

```python
!pip install numpy
import numpy as np
```

```python
# Generate Random Sales data using numpy
np.random.seed(15) # use a seed to get the same random data

# 4 Quarters
quarters = ['Q1', 'Q2', 'Q3', 'Q4']

# Define the sales data of male and female randomly
sales_female = np.random.randint(15, 50, size=len(quarters))
sales_male = np.random.randint(15, 50, size=len(quarters))

# Create a new figure with the size of 20x20 inches
plt.figure(figsize=(20,20))

# Plot the sales data with styling
plt.plot(quarters, sales_female, marker='o', color='red', label='Female')
plt.plot(quarters, sales_male, marker='s', color='blue', label='Male')

# Add labels
plt.xlabel('Quarter')
plt.ylabel('Sales')

# Add title
plt.title('Product Sales of Company X by Gender')

# Show legend
plt.legend()

# Add grid
plt.grid(True)

# Set the quarters to the x-axis
plt.xticks(quarters)

# Adjust the y-axis accordingly
plt.ylim(0, max(max(sales_female), max(sales_male)) + 20)

# Display graph
plt.show()
```
